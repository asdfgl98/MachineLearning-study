#### Age 결측치 채우기
- 다른 컬럼들과의 상관 관계를 사용해서 결측치 채우기

```python
# 수치형 컬럼들간의 상관관계를 나타내는 함수 : corr()
# corr() 사용시 흔히 발생하는 문제
# 1. 누락된 값 : 있는 경우 NaN 값의 행렬 반환
# 2. 숫자가 아닌 열이나 데이터 유형이 포함되어있는 경우 이를 무시
# 3. 가지고온 데이터 프레임에 불균형한 수가 포함되어 있을 경우
#   -> 상관계수 의미가 없어짐

# 해결방법 
# 1. 누락된 값 삭제 : corr() 적용전에 dropna() 함수를 이용해서 누락된 값을 포함하고있는 행 삭제
# 2. 숫자가 아닌 열 삭제 : corr() 적용전에  select_dtypes() 함수를 사용해서 숫자열만 선택
#      select_dtypes(include=[], exclude=[]) include = 포함 , exclude = 제외
# 3. 데이터를 표준화 : scikit-learn 라이브러리의 StandardScaler를 사용하여 데이터 표준화
train.select_dtypes(include=['int64', 'float64']).corr()


# -1 ~ 1 사이의 숫자로 표현
# -1 에 가까울수록 반비례 : 나이가 많을수록 객실 등급은 1등급에 가깝다.
# 1에 가까울수록 비례

# 수치앞의 부호는 비례, 반비례인지 판단 기준
# 부호뒤의 숫자의 절대값이 1에 가까울수록 서로 영향을 준다고 판단

# 형제자매/배우자수, 객실등급이 나이와 상관이 있을 것이다.
```

```python
# Age 컬럼을 Pclass컬럼과 비교해서 결측치 채우기

train[['Pclass','Age']] # --> Pclass 별로 묶어서 중앙값 확인
train[['Pclass','Age']].groupby(by='Pclass').median() # Pclass 별로 탑승한 승객 나이의 중앙값 
#                                                      (Pclass 1,2,3 별로 묶어서 그 중에 중앙값 .median으로 출력)
# Pclass = 1, 2, 3 : 숫자자료이지만 등급을 나타내는 범주형 자료
```

```python
# Age 컬럼을 Pclass 먼저추가하고, Sex 컬럼 추가하기
# Pclass컬럼 별로 탑승한 승객 나이의 중앙값을 활용 ==> 결측치 채우기

# 3개의 컬럼을 보기위해 Age_table로 저장 (테이블 형태)
age_table = train[['Pclass', 'Sex', 'Age']].groupby(by=['Pclass','Sex']).median()
age_table
```

```python
train.iloc[0]
```

```python
# 하나의 행을 가지고와서
# Age가 결측치가 아닌 경우  --> 원래 가지고 있던 값을 그대로 사용
# Age가 결측치라면 --> Plcass와 Sex를 확인해서 age_table에서 해당하는 중앙값으로 채우기

# 꺼내오는 하나의 행 정보 : 넘파이 배열형태
import numpy as np

# 함수정의
# 함수를 만들 때 주의할점
# 행단위로 데이터가 들어올 것으로 예상하고 만들기

def fill_age(person):
    if np.isnan(person['Age']) : # 꺼내온 정보에서 Age 컬럼이 결측치라면 실행
        return age_table.loc[person['Pclass'], person['Sex']]['Age']
    # 결측치라면 age_table의 Pclass와 Sex를 확인하고 Age값을 적용
    else: # Age 컬럼이 결측치가 아님
        return person['Age'] # 가지고있는 값을 사용
    
```

```python
# 2. 승객의 정보를 확인하고 함수 적용 ==> apply() 
# apply() : 행과 열 단위의 복잡한 계산을 할 때 사용하는 함수
age_table.loc[3, 'male']['Age']

train.apply(fill_age, axis=1)
```

```python
train['Age'] = train.apply(fill_age, axis=1)
test['Age'] = test.apply(fill_age, axis=1)
```

- 구간별로 구분된 수치형 데이터를 범주형(문자) 데이터로 변경 -> Binning 함수

##### Name
- 1. Name 데이터들 사이에서 공통적인 특징 찾기
- 2. 찾은 특징들로 데이터를 다시 정리

- Name 데이터의 공통점 : 이름과 성 사이에 호칭(Mr, Mrs)이 존재
- Name이라는 글자 자체는 중요하지 않고 호칭으로 그룹화 한 것들을 숫자로 바꿈

```python
train['Name']
```

```python
# 이름 중간에 있는 호칭만 가져오기
train['Name'][0]

# 1. 쉼표(,)를 기준으로 데이터를 분할하고 뒤에있는 데이터를 사용 # split() : 데이터를 특정 문자를 기준으로 분리
# 2. 마침표(.)를 기준으로 데이터를 분할하고 앞에있는 데이터를 사용
train['Name'][0].split(',')[1].split('.')[0].strip()  # strip() : 앞뒤 공백을 제거하는 함수
```

```python
# 중간 호칭을 가져오는 함수
def split_name(name) :
    return name.split(',')[1].split('.')[0].strip()
    
```

```python
# apply 함수 적용
train['Name'] = train['Name'].apply(split_name)
test['Name'] = test['Name'].apply(split_name)
```

### 성능(점수) 개선하기
- 하이퍼파라미터 튜닝
- 다른 모델 사용하기

####  하이퍼 파라미터 튜닝하기
- 모델이 학습용 데이터에 과대적합이 걸려있으니 이를 해소해보자
- 과정 : 하이퍼파라미터 튜닝 -> 학습 -> train, test 확인
- 최종 : train과 test의 score의 차이가 크지 않다면 ==> 잘 만들어진 모델

```python
# 5. 하이퍼 파라미터 튜닝
tree = DecisionTreeClassifier(max_depth=5)
# 6. 학습
# tree.fit(X_train,y_train) -> 교차검증 시 내부에서 학습이 이루어지기때문에 따로 학습을 하지않아도 됨
# 교차검증으로 평가
result = cross_val_score(tree, X_train, y_train, cv=5).mean()
print(f'max_depth : 5, score : {result}')
```

```python
for i in range(1, 11):
    tree = DecisionTreeClassifier(max_depth=i)
    result = cross_val_score(tree, X_train, y_train, cv=5).mean()
    print(f'max_depth : {i}, score : {result}')
```
